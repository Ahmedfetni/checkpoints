```python
from sklearn.linear_model import LogisticRegression # modele 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score # for testing performance 
```

```python
cols =  list(data.keys())
cols.pop()
x = data[cols]
y = data.Survived # target

cols

```

```python
# spliting the data 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0) 

model = LogisticRegression()  

model.fit(x_train,y_train) 
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
classification_report(y_pred,y_test)
```

# KNN

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull()
```

```python
#cleaning the data set 
data = data.drop('Name', axis=1,)
data = data.drop('Ticket', axis=1,)
data = data.drop('Fare', axis=1,)
data = data.drop('Cabin', axis=1,)
data['Family'] = data['SibSp'] + data['Parch'] + 1
```

```python
data["Age"] = data["Age"].fillna(data["Age"].median())
```

```python
data['Embarked'].value_counts()
```

```python
print(data["Embarked"].mode())
```

```python
data["Embarked"] = data["Embarked"].fillna("S")
```

```python
data["Embarked"].unique()
```

```python
data.Embarked.replace(['S', 'C', 'Q'], [1, 2, 3], inplace=True)
```

```python
data.Sex.replace(['male', 'female'], [1,0], inplace=True)
```

```python
data.isnull()
```

```python
y = np.array(data.filter(['Survived'], axis=1))
X=np.array(data.filter(['Pclass','Sex','Embarked','Family','Age'], axis=1))
```

```python
X_1, X_test, y_1, y_test = train_test_split(X,y, test_size=0.3)
```

```python

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_1, y_1)
pred = knn.predict(X_test)
confusion_matrix(pred,y_test)
#classification_report(pred,y_cv)
```
